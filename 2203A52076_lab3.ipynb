```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import random as rd
a=pd.read_csv('/content/heart.csv')
x=a.drop("target",axis=1)
y=a["target"]
scaler=MinMaxScaler()
scaler.fit(x)
x_scaled=scaler.transform(x)
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42)
f=x_train.shape[1]
w=[]
for _ in range(f):
    weight=rd.random()
    w.append(weight)
bias=rd.random()
def perceptron(x):
    z = (
        x[0]*w[0]+
        x[1]*w[1]+
        x[2]*w[2]+
        x[3]*w[3]+
        x[4]*w[4]+
        x[5]*w[5]+
        x[6]*w[6]+
        x[7]*w[7]+
        x[8]*w[8]+
        x[9]*w[9]+
        x[10]*w[10]+
        x[11]*w[11]+
        x[12]*w[12]+
        bias
    )
    return 1/(1+np.exp(-z))
y_p=[]
for x in x_train:
    if perceptron(x)>=0.5:
        y_p.append(1)
    else:
        y_p.append(0)

c_p=np.sum(y_p==y_train)
total=len(y_train)
acc=c_p / total
print(f'Accuracy:{acc:.2f}')
```
